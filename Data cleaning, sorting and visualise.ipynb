```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student's_Score.csv")
print(df.head())
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis=1)
print(df)
```

# Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender", palette=['pink', 'skyblue'])
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution", color = "Maroon")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of female in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean' })
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
#sns.heatmap(gb,cmap='viridis')
plt.title("Relationship between Parent's Education and Student's Score ", color="Purple")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their students
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean' })
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
#sns.heatmap(gb,cmap='viridis')
plt.title("Student's Score due to their Parent's Marital Status ", color="Brown")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the 
#student's score due to their parent's marital status
```

# Distribution of Ethnic Group

```python
groupA =df.loc[(df["EthnicGroup"]== "group A")].count()
print(groupA)
```

```python
groupA =df.loc[(df['EthnicGroup']== "group A")].count()
groupB =df.loc[(df['EthnicGroup']== "group B")].count()
groupC =df.loc[(df['EthnicGroup']== "group C")].count()
groupD =df.loc[(df['EthnicGroup']== "group D")].count()
groupE =df.loc[(df['EthnicGroup']== "group E")].count()
l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)

plt.pie(mlist, labels =l, autopct ="%1.2f%%", colors=sns.color_palette('Set3'))
plt.title("Distribution of Ethnic Groups", color="Brown")
plt.show()
```
